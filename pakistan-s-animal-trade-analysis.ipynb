# Correlation Between Columns

```python
correlation_matrix = imports[['Quantity','Value']].corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot= True, cmap='coolwarm')
plt.show()
```

```python
correlation_matrix = imports[['Quantity','Value','Year']].corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot= True, cmap='coolwarm')
plt.show()
```
